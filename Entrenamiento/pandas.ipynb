```python
import pandas as pd

value_n_list = ["Andrew", "Phillip", "John"]
print(value_n_list)

value_w_list = [50,     78,     91.5]
value_h_list = [1.76,   1.87,   1.69]

data_dict = {"name": value_n_list,
             "weight":value_w_list,
             "height":value_h_list
}

print(data_dict)

df = pd.DataFrame(data_dict)
print(df)
```
